```python
import gzip
import json
import re
import os
import csv
import ast
import sys
import numpy as np
import pandas as pd
```

```python
DIR = './'
```

```python
def load_data(file_name):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            


            # break if reaches the 100th line
#             if (head is not None) and (count > head):
#                 break
   
    return data
```

```python
reviews = load_data(os.path.join(DIR, 'goodreads_reviews_spoiler.json.gz'))
```

```python
reviews
```

```python
field_names = ['user_id', 'timestamp','review_sentences','rating','has_spoiler','book_id','review_id']
```

```python
with open('book_spoilers.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(reviews)
```
